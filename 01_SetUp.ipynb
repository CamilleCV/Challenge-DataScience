## SetUp

```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', 100)
```

```python
uri = 'https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json'
data = pd.read_json(
    path_or_buf = uri,
    orient = 'columns'
)

data.head()
```

## Conhecendo o DataFrame

Os dados estão em *JSON*, contendo dicionários dentro das colunas. 

### Verificando os Dicionários

### Transformando os dados JSON em Tabela

```python
dados_customer = pd.json_normalize(data['customer'])
dados_customer.head()
```

```python
dados_phone = pd.json_normalize(data['phone'])
dados_phone.head()
```

```python
dados_internet = pd.json_normalize(data['internet'])
dados_internet.head()
```

```python
dados_account = pd.json_normalize(data['account'])
dados_account.head()
```

### Concatenando as Tabelas

```python
df = data[['customerID','Churn']]

df = pd.concat([df, dados_customer, dados_phone, dados_internet, dados_account], axis = 1)

print('O DataFrame original contém %d linhas.' % data.shape[0])
print('O DataFrame concatenado contém %d linhas.' % df.shape[0])
```

### Analisando o DataFrame

```python
df.head()
```

```python
df.info()
```

A coluna *Charges.Total* está como objeto... <p>
A primeira tratativa foi utilizando o *.astype(float)*, mas o comando estava retornando um erro... Desse modo, outra ação foi tomada:

```python
df['Charges Total'] = pd.to_numeric(df['Charges.Total'], errors='coerce')

df['Charges Total'].info()
```

Com essa análise, percebeu-se que, das 7267 entradas, apenas 7256 continham informações que poderiam ser transformadas em float. <p>
Para compreender melhor, esses dados foram inseridos em uma nova coluna, filtrados e visualizados no próximo dataframe.

```python
df[df['Charges Total'].isnull()]
```

Como é possível verificar, as informações não numéricas estavam vazias, mas não apareciam como *NaN* no DataFrame. <p>
Por serem apenas 11 informações, optou-se por excluí-los da base de dados.

```python
df = df.dropna()
df['Charges.Total'] = df['Charges Total']
df.drop('Charges Total', axis = 1, inplace=True)
df.head()
```
